```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"Z:\MMS\lab-work-m1-AA-maker-dev-main\Lab-04\Image\1004016.jpg")
print(image)

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```
